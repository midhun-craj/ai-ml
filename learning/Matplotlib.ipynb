```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
import matplotlib.font_manager as fm

font_names = sorted(set(f.name for f in fm.fontManager.ttflist))

print(font_names)
```

```python
# plt.plot([1, 2, 3], [2, 4, 6])

# x = [1, 2, 3]
# y = [2, 4, 6]

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize you graph (dpi specifies pixels per inch. when saving probably should use 300 or a higher number if possible)
plt.figure(figsize=(5, 3), dpi=100) # Make the dpi higher to keep the graph from being pixelated


# Line 1

# Keyword argument notation
# plt.plot(x, y, label='2x', color='red', linewidth='2', marker='.', markersize='12', markeredgecolor='black', linestyle='--')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r.--',label='2x')


# Line 2

# Select intervals we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'y.-', label='x^2')

# Plot remainder of graph as dashed line
plt.plot(x2[5:], x2[5:]**2, 'b.--')

# Add a title (specify font parameters with fontdict)
plt.title("Our First Graph!", fontdict={'fontname': 'DejaVu Serif', 'fontsize': 20})

# x & y labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# x & y axis Tickmarks (scale of your graph)
plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8])

plt.legend()

# Save figure (dpi 600 or even 300 is better when saving so graph has high resolution)
plt.savefig('MyGraph', dpi=600)

# Show plot
plt.show() # To remove this line from the output "[<matplotlib.lines.Line2D at 0x7e4c4c259150>]"
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 3, 2]

bars = plt.bar(labels, values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('.')

patterns = ['/', '*', '.']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.title("Our First Bar Chart", fontdict={'fontsize': 20})

plt.figure(figsize=(6, 4), dpi=600)

plt.show() # To remove this line from the output "<BarContainer object of 3 artists>"
```
